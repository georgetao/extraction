```python
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
import json
```

```python
EPA_data_path = '../data/EPADataset.tsv'
```

```python
epa = pd.read_csv(EPA_data_path, sep='\t')
```

```python
epa['record'] = epa['Json'].map(json.loads)
df = pd.concat(epa['record'].map(json_normalize).tolist()).reset_index(drop=True)
```

```python
for rec in df.itertuples():
    print(10*'=')
    print('Message:')
    print(rec.Message)
    print('\nTaskSentence:')
    print(rec.TaskSentence)
    print()
```
